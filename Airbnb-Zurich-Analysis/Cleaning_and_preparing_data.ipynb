```python
import pandas as pd
import ast 
```

## Loading data

```python
# Loading main data files: listings, reviews, neighborhoods, and calendar
listings = pd.read_csv("listings.csv")
reviews = pd.read_csv("reviews.csv")
neighbourhoods = pd.read_csv("neighbourhoods.csv")
calendar = pd.read_csv("calendar.csv")
```

## Listings

```python
pd.set_option('display.max_columns', 75) 
listings
```

```python
# Getting general information about data structure: types, missing values, row count
listings.info(verbose=True, show_counts=True)
```

```python
listings[listings['id'].isna() | (listings['id'] == 0)]
```

```python
# Exploratory duplicates
duplicate_ids = listings[listings.duplicated(subset='id', keep=False)]
duplicate_ids
```

```python
# Removing duplicates,keeping only the first occurrence for each unique ID
listings = listings.drop_duplicates(subset='id', keep='first')
```

```python
#Removing records with missing IDs and columns consisting entirely of NULL values
listings = listings[listings['id'].notnull() & (listings['id'] != 0)]
listings = listings.dropna(axis=1, how='all')
```

```python
# Logically fill in missing values for bedrooms and beds
listings['bedrooms'] = listings['bedrooms'].fillna(listings['accommodates'].apply(lambda x: max(1, round(x/2))))
listings['beds'] = listings['beds'].fillna(listings['bedrooms'])
```

```python
 #Adding useful business metrics for deeper analysis
listings['price_per_person'] = listings['price'] / listings['accommodates']
listings['price_per_bedroom'] = listings['price'] / listings['bedrooms']
listings['price_category'] = pd.qcut(listings['price'], 4, labels=['Budget', 'Economy', 'Mid-Range', 'Premium'])
```

```python
#Converting JSON string structure to Python list for easier analysis
listings['amenities'] = listings['amenities'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else [])
```

```python
# Converting coordinates from string format with commas to numerical format
listings['latitude'] = listings['latitude'].str.replace(',', '.')
listings['longitude'] = listings['longitude'].str.replace(',', '.')

columns_to_convert = [
    'minimum_nights_avg_ntm',
    'maximum_nights_avg_ntm',
    'review_scores_rating',
    'review_scores_accuracy',
    'review_scores_cleanliness',
    'review_scores_checkin',
    'review_scores_communication',
    'review_scores_location',
    'review_scores_value',
    'reviews_per_month',
    'latitude',
    'longitude'
]

for col in columns_to_convert:
    if col in listings.columns:
        listings[col] = pd.to_numeric(listings[col], errors='coerce')

# Converting from float to integer
integer_columns = ['bedrooms', 'beds', 
                   'minimum_nights', 'maximum_nights',
                   'minimum_minimum_nights', 'maximum_minimum_nights',
                   'minimum_maximum_nights', 'maximum_maximum_nights',
                   'availability_30', 'availability_60', 'availability_90', 'availability_365',
                   'number_of_reviews', 'number_of_reviews_ltm', 'number_of_reviews_l30d',
                   'calculated_host_listings_count', 
                   'calculated_host_listings_count_entire_homes',
                   'calculated_host_listings_count_private_rooms',
                   'calculated_host_listings_count_shared_rooms']

for col in integer_columns:
    if col in listings.columns:
        # Check that the values can be safely converted to integer
        if listings[col].notnull().all() and (listings[col] % 1 == 0).all():
            listings[col] = listings[col].astype(int)
        else:
            listings[col] = listings[col].astype('Int64')
```

```python
listings['id'] = listings['id'].astype('int64')
```

```python
listings.info(verbose=True, show_counts=True)
```

```python
# Converting from object to datetime
date_cols = ['last_scraped', 'host_since', 'first_review', 'last_review', 'calendar_last_scraped']
for col in date_cols:
    if col in listings.columns:
        listings[col] = pd.to_datetime(listings[col], dayfirst=True)
```

```python
listings.info(verbose=True, show_counts=True)
```

```python
#Replace '%' to ' ' for better evaluating and change type from sting to float 
percent_columns = ['host_response_rate', 'host_acceptance_rate']

for col in percent_columns:
    if col in listings.columns:
        listings[col] = listings[col].str.replace('%','', regex=False)
        listings[col] = pd.to_numeric(listings[col], errors='coerce')
listings.rename(columns={
    'host_response_rate': 'host_response_rate (%)',
    'host_acceptance_rate': 'host_acceptance_rate (%)'
}, inplace =True)
```

```python
# Looking at unique values
print(listings['host_is_superhost'].unique())
```

```python
# Converingt from object to bool
listings['host_is_superhost'] = listings['host_is_superhost'].astype(str).map({'True': True, 'False': False, 'nan': False})
```

```python
listings.info(verbose=True, show_counts=True)
```

## calendar

```python
calendar.info(verbose=True,show_counts=True)
```

```python
# Converting from object to float
columns_to_correct = ['price','adjusted_price']
for col in columns_to_correct:
    calendar[col] = calendar[col].str.replace('$', '', regex = False)
    calendar[col] = pd.to_numeric(calendar[col],errors = 'coerce')
calendar.rename(columns={'price' : 'price $','adjusted_price': 'adjusted_price $'},inplace =True)
```

```python
# Converingt from object to bool
calendar['available'] = calendar['available'].map({'t': True, 'f': False})
```
